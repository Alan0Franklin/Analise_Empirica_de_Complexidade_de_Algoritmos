```python
df = pd.read_csv('Registro_-_Tempos_de_Execucao.csv', sep = ';')
df = df.iloc[1:]
df
```

> ***Complexidades Teóricas para o Caso Médio de cada Algoritmo***

```python
# Complexidades Teóricas
O_n2 = df['n']**2                         # Bubble Sort
O_nlogn = df['n'] * np.log2(df['n'])      # Merge Sort
O_n = df['n']                             # Busca Sequencial
O_logn = np.log2(df['n'])                 # Busca Binária
```

> ***Geração do Gráfico do 'Tempo de Execução dos Algoritmos × Quantidade de Entradas', com a adição das complexidades teóricas do caso médio de cada algoritmo***

```python
plt.figure(figsize=(10, 6))

# Plotagem dos Dados Empíricos de cada Algoritmo
plt.plot(df['n'], df["Bubble Sort"]/df["Bubble Sort"].max(), label="Bubble Sort (empírico)", color="#4878CF")
plt.plot(df['n'], df["Merge Sort"]/df["Merge Sort"].max(), label="Merge Sort (empírico)", color="#EAEA60")
plt.plot(df['n'], df["Busca Sequencial"]/df["Busca Sequencial"].max(), label="Busca Sequencial (empírico)", color="#3CBA54")
plt.plot(df['n'], df["Busca Binaria"]/df["Busca Binaria"].max(), label="Busca Binaria (empírico)", color="#E24A33")

# Plotagem das Complexidades Teóricas
plt.plot(df['n'], O_n2/O_n2.max(), label="O(n²) teórico", color="#4C72B0", linestyle = '--')
plt.plot(df['n'], O_nlogn/O_nlogn.max(), label="O(n ⋅ log(n)) teórico", color="#CCBD4A", linestyle = '--')
plt.plot(df['n'], O_n/O_n.max(), label="O(n) teórico", color="#55A868", linestyle = '--')
plt.plot(df['n'], (O_logn - O_logn.min())/O_logn.max() , label="O(log(n)) teórico", color="#C44E52", linestyle = '--')

plt.title('Tempo de Execução vs Quantidade de Entradas')
plt.xlabel('Quantidade de Entradas (n)')
plt.xlim(0, 100000)
plt.ylabel('Tempo de Execução normalizado')
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
